### --> Droping rows/coloumns with null values

### --> Replacing null values with non-null values

```python
df1 = df.fillna(0)
df1.info()
```

### --> Visualization

```python
#LIFE EXPECTANCY WRT COUNTRY
x = df1['Country']
y = df1['Life expectancy ']

f = plt.figure()
f.set_figwidth(40)
f.set_figheight(20)

plt.xlabel('Counrties', fontsize = '50')
plt.ylabel('Life expectancy', fontsize = '50')

plt.plot(x,y)
```

```python
#DISEASES WRT YEARS
import warnings
warnings.filterwarnings('ignore')

x = df['Year']
y = df['Hepatitis B']
z = df['Polio']

f = plt.figure()
f.set_figwidth(40)
f.set_figheight(20)

plt.ylabel('Year', fontsize = '50')
plt.xlabel('Diseases', fontsize = '50')

plt.bar(y + 0.0, x, color = 'blue', width = 0.5)
plt.bar(z + 0.5, x, color = 'yellow', width = 0.5)

plt.ylim(2000,2017)
plt.legend(['Hepatitis B', 'Polio'])
```
